```python
import pandas as pd
```

```python
# read in tounrmanel model csv

df = pd.read_csv('data/tournament_model_ml.csv')
df.head()

df_train = df[df['season'] <= 2022]
df_test = df[df['season'] > 2022]
```

```python
def baseline(df):
    # copy to avoid SettingWithCopyWarning on slices
    df = df.copy()
    # win_label is 1 when the first team is stronger; seed_diff is negative when the first team is the better seed.
    df['pred'] = 0
    df.loc[df['seed_diff'] < 0, 'pred'] = 1
    df.loc[df['seed_diff'] == 0, 'pred'] = (df.loc[df['seed_diff'] == 0, 'SRS_diff'] > 0).astype(int)
    return df
```

```python
results = baseline(df_test)
#results = results[['pred', 'result']]
#results.head()
# # calculate accuracy
accuracy = (results['pred'] == results['win_label']).mean()
print(f'Baseline accuracy: {accuracy:.4f}')

# make a confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(results['win_label'], results['pred'])
print('Confusion Matrix:')
print(cm)
```
