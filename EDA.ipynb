# EDA for March Madness Data
## Summary Statistics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("data/tournament_model_ml.csv")

# Quick glance
df.head()
```

```python
# Categorical variables
categorical_vars = ["win_label", "season"]

# Numeric variables
numeric_vars = [col for col in df.columns if col not in categorical_vars]

numeric_vars, categorical_vars
```

```python
numeric_summary = df[numeric_vars].agg(
    ["count", "mean", "std"]
).T

numeric_summary
```

```python
five_number_summary = df[numeric_vars].describe().T[
    ["min", "25%", "50%", "75%", "max"]
]

five_number_summary
```

```python
numeric_full_summary = numeric_summary.join(five_number_summary)
numeric_full_summary
```

```python
for col in categorical_vars:
    print(f"\n=== {col.upper()} ===")
    print("Sample Size:", df[col].count())
    print(df[col].value_counts())
```

```python
correlation_matrix = df[numeric_vars].corr()
correlation_matrix
```

```python


corr_table = (
    df[numeric_vars + ["win_label"]]
    .corr()
    [["win_label"]]
    .rename(columns={"win_label": "Correlation with Win"})
    .sort_values("Correlation with Win", ascending=False)
    .round(3)
)

corr_table
```
